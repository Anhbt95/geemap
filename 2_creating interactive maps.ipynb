# Adding basemaps


```python
# basemap hybrid
Map = geemap.Map(basemap='HYBRID')
#Map
# add OpenTopoMap (can turn on and off)
Map.add_basemap('OpenTopoMap')
Map.add_basemap("Strava.Water")
Map
```

```python
# check basemaps in ggmap (148 basemaps)
for basemap in geemap.basemaps.keys():
    print(basemap)
```

```python
# add wms tiles - NLCD 2019
## first need to allow to import
from google.colab import output
output.enable_custom_widget_manager()

Map = geemap.Map(center=[40, -100], zoom=4)
url = 'https://www.mrlc.gov/geoserver/mrlc_display/NLCD_2019_Land_Cover_L48/wms?'
Map.add_wms_layer(
    url=url,
    layers='NLCD_2019_Land_Cover_L48',
    name='NLCD 2019',
    format='image/png',
    attribution='MRLC',
    transparent=True,
)
Map

```
